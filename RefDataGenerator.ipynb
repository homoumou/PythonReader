```python
import pandas as pd
## sheet files
xlsx = pd.ExcelFile('BPS Anatomical body structure list.xlsx')
for sheetName in xlsx.sheet_names:
  # print(sheetName)
  df = pd.read_excel(xlsx, sheetName)
  for index, row in df.iterrows():
    try:
      Code = row['Code']
      Snowed = row['Concept ID']
      Text = row['OMNI Term']
      HasSide = 'true' if row['Side'] == 'y' else 'false'
      WeightBearing = 'true' if row['WB'] == 'y' else 'false'
      IsActive = 'true' if row['isActive'] == 1 else 'false'
      print( Code +': {Text: "'+ Text +'", Snowed: '+ str(int(Snowed)) +', HasSide: '+ HasSide +', WeightBearing: '+ WeightBearing +', SortOrder: '+ str(index + 1)+ ', IsActive: '+ IsActive+ ' }' )
    except Exception as e:
        print(e)
        pass
```
